**<h1>Korrelationsmatrix:**

```python
correlation_matrix = data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

<h3>Wir sehen, dass:


- Schokolade den **größten positiven Einfluss** auf die Beliebtheit hat
- Fruchtige Süßigkeiten den **größten negativen Einfluss** auf die Beliebtheit haben

- **Riegel** i.d.R aus Schokolade und/oder Nougat bestehen und selten fruchtig oder eine Multivariante sind
- Zucker **keinen signifikaten Einfluss** auf die Eigenschaften und somit auch nicht auf die Beliebtheit hat

**<h1>Analyse der einzelnen Merkmalsträger:**

- **Annahme Zucker**:
  Anhand der Korrelationsmatrix lässt sich erkennen, dass der Zuckeranteil keinen starken Einfluss auf die einzelnen Merkmale hat. Daher wird der Zuckeranteil nicht berücksichtigt.
- **Annahme Preis**:
  Es wird angenommen, dass die Befragten nicht für die Süßigkeiten zahlen mussten, daher wird zur Herleitung der Beliebtheit der Preis nicht berücksichtigt.
  Da der Preis für die Herstellung jedoch von Relevanz ist, wird dieser bei der Analyse der einzelnen Merkmale betrachtet und dargestellt.

***<h3>Schokolade:***

```python
plt.figure(figsize=(30,10))

plt.suptitle("Schokolade Plots", fontsize=20)

plt.subplot(1,3,1)
plt.title("Wining precentage mit & ohne Schokolade",fontsize=15)
sns.boxplot(data=data,x="chocolate",y="winpercent",color="brown")

plt.subplot(1,3,2)
plt.title("Totaler Share von Süßigkeiten mit & ohne Schokolade",fontsize=15)
plt.pie(data.chocolate.value_counts(), autopct='%1.1f%%', labels=["Ohne Schokolade","Mit Schokolade"])

plt.subplot(1,3,3)
plt.title("Price percentage mit und ohne Schokolade",fontsize=15)
sns.boxplot(y="pricepercent", x="chocolate", data=data)


plt.show()
```

```python
Schoko_win = data[['chocolate', 'winpercent']]
Schoko_win_avg = Schoko_win[Schoko_win['chocolate'] == 1].groupby(['chocolate'], as_index=False).mean()
Schoko_win_avg
```

<h3>Die dargestellen Graphen zeigen, dass Schokolade:</h3>

- **einen stark positiven Einfluss** auf die Beliebtheit haben und im Durchschnitt zu 61% von den Befragten bevorzugt werden
- knapp 45% aller Süßigkeiten Schokolade enthalten
- teurer ist als alternative Produkte

```python
print('p-value = {0:.10%}'.format(stats.ttest_ind(data[data.chocolate == 0].winpercent, 
                                               data[data.chocolate == 1].winpercent)[1]))
```

<h4>Der p-Wert bestätigt, dass Schokolade einen statistisch signifikanten Einfluss auf die Beliebtheit hat.

**<h1>Multi-lineares Regressionsmodell zur Ermittlung der einzelnen Koeffizienten:**

<h4>Nach der Analyse der einzelnen Merkmale, schauen wir nun inwiefern eine Kombination aus einzelnen Merkmalen einen positiven Einfluss auf die Beliebtheit haben kann.

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm
```

```python
Z = data[['chocolate', 'fruity', 'peanutyalmondy', 'crispedricewafer', 'caramel', 'nougat', 'bar', 'hard', 'pluribus']]
```

```python
lm.fit(Z, data['winpercent'])
```

```python
lm.intercept_
```

<h4>Der Y-Achsen Schnittpunkt beträgt 35,02</h4>

```python
lm.coef_
```

<h4>Der dargestellte Array gibt die Koeffizienten der einzelnen Merkmale an. Diese geben uns Auskunft darüber, wie sehr sich der Wert von "winpercent" ändert, wenn eine der Eingangsvariablen zutrifft.

```python
# Find the R^2
print('The R-square is: ', lm.score(Z, data['winpercent']))
```

<h4>Das Bestimmtheitsmaß des Modells beträgt 51,5%. Wir können anhand des Modells also 51,5% der Varianz der abhängigen "winpercent" Variable anhand der unabhängigen Variablen erklären. Dies ist eine moderate bis gute Leistung, für unseren Zweck jedoch ausreichend.</h4>

```python
Y_predict_multifit = lm.predict(Z)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(data['winpercent'], Y_predict_multifit)
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```

<h4>Der RMSE des Modells beträgt 10,19. Das beudetet, dass im Durchschnitt unser Modell die Zielvariable "winpercent" um etwa 10 Prozentpunkte verfehlt.</h4>
